```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import os
from tensorflow import keras
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, RandomFlip, RandomRotation
from tensorflow.keras.models import Model
from tensorflow.keras.applications import EfficientNetB0
import tarfile
from PIL import Image
```

```python

import zipfile

# Unzip the file containing the cat and dog photos
with zipfile.ZipFile('cats_n_dogs.zip', 'r') as zip_ref:
    zip_ref.extractall('catdog')
```

```python
# Check the contents
print(os.listdir('catdog'))  # Should show ['cats', 'dogs'] or similar
print(f"Number of cat images: {len(os.listdir('cats_n_dogs/cats_n_dogs/Cat'))}")
print(f"Number of dog images: {len(os.listdir('cats_n_dogs/cats_n_dogs/Dog'))}")
```

```python
# Define paths and parameters
data_dir = 'cats_n_dogs/cats_n_dogs'
IMG_SIZE = 224 # EfficientNetB0 uses 224x224 images
BATCH_SIZE = 32
```

```python

def check_images(directory):
    """Check for corrupted images in the dataset"""
    corrupted_files = []
    
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
                filepath = os.path.join(root, file)
                try:
                    img = Image.open(filepath)
                    img.verify()  # Verify it's a valid image
                except Exception as e:
                    print(f"Corrupted: {filepath} - Error: {e}")
                    corrupted_files.append(filepath)
    
    return corrupted_files

# Check for corrupted images
print("Checking for corrupted images...")
corrupted = check_images(data_dir)

if corrupted:
    print(f"\nFound {len(corrupted)} corrupted image(s)")
    print("Removing corrupted files...")
    for file in corrupted:
        os.remove(file)
        print(f"Removed: {file}")
else:
    print("No corrupted images found!")
```

```python
# Create training and validation datasets
# We'll use 80% of the data for training and 20% for validation
train_dataset = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE
)
```

```python
validation_dataset = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE
)
```

```python
# Check the class names
# Get the class names (these will be the names of your subfolders)
class_names = train_dataset.class_names
print(f"Classes found: {class_names}")

# Check the shapes
for images, labels in train_dataset.take(1):
    print(f"Image batch shape: {images.shape}")
    print(f"Label batch shape: {labels.shape}")
```

```python
# Add this to ignore errors during iteration
train_dataset = train_dataset.apply(tf.data.experimental.ignore_errors())
validation_dataset = validation_dataset.apply(tf.data.experimental.ignore_errors())
```

```python

# Configure the dataset for performance
AUTOTUNE = tf.data.AUTOTUNE
train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
validation_dataset = validation_dataset.prefetch(buffer_size=AUTOTUNE)
```

```python
plt.figure(figsize=(10, 10))
for images, labels in train_dataset.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
plt.show()
```

```python
# Add data augmentation layers to prevent overfitting
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal'),
    tf.keras.layers.RandomRotation(0.2),
])
```

```python
normalization_layer = tf.keras.layers.Rescaling(1./255)
normalized_ds = train_dataset.map(lambda x, y: (normalization_layer(x), y))
image_batch, labels_batch = next(iter(normalized_ds))
first_image = image_batch[0]
# Notice the pixel values are now in `[0,1]`.
print(np.min(first_image), np.max(first_image))
```

```python
# Apply BOTH augmentation and normalization to training data
train_dataset = train_dataset.map(
    lambda x, y: (data_augmentation(normalization_layer(x), training=True), y),
    num_parallel_calls=AUTOTUNE
)

# Apply ONLY normalization to validation data (no augmentation)
validation_dataset = validation_dataset.map(
    lambda x, y: (normalization_layer(x), y),
    num_parallel_calls=AUTOTUNE
)

print("Augmentation and normalization applied!")
```

```python
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape =[224, 224, 3]),  # Added 3 for RGB channels
    keras.layers.Dense(300, activation="relu"),
    keras.layers.Dense(100, activation="relu"),
    keras.layers.Dense(2, activation="softmax")
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',  # Use this for integer labels
    metrics=['accuracy']
)

# Check the model structure
model.summary()
```

```python
# Train the model
history = model.fit(
    train_dataset,
    validation_data=validation_dataset,
    epochs=10  # Adjust as needed
)
```

```python
# Extract history
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(10)

# Plot Training and Validation Accuracy
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

# Plot Training and Validation Loss
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')

plt.suptitle('Model Performance Metrics', fontsize=16)
plt.show()
```
